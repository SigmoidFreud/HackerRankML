```python
featuresets = [['English', 'Physics', 'Chemistry', 'Mathematics', 'ComputerScience'],
               ['English', 'Physics', 'Chemistry', 'Mathematics', 'PhysicalEducation'],    
               ['English', 'Physics', 'Chemistry', 'Mathematics', 'Economics'],    
               ['English', 'Physics', 'Chemistry', 'Mathematics', 'Biology'],  
               ['English', 'Economics', 'Accountancy', 'Mathematics', 'BusinessStudies']]

df_breakdown = []

def gen_breakdown(featureset):
    dataframe = df
    for colname in df[featureset]:
        dataframe = dataframe[dataframe[colname]>0]
    return dataframe

for featureset in featuresets:
    df_breakdown.append(gen_breakdown(featureset)[['serial']+featureset])

df_breakdown
```

```python
for frame in df_breakdown:
    frame['is_train'] = np.random.uniform(0, 1, len(frame)) <= .75
    
train = []
test = []

for df in df_breakdown: # creates a set of test and train data
    train.append(df[df['is_train']==True].drop('is_train',axis=1)) 
    test.append(df[df['is_train']==False].drop('is_train',axis=1))
    

test
```

```python
corrmat = []
for df in enumerate(df_breakdown):
    corrmat.append(df[1][featuresets[df[0]]].corr())
corrmat[3]
```

```python
# compute optimal number of features for each classifier
# changing the parameters will likely change results
from sklearn.svm import SVC
from sklearn.cross_validation import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.datasets import make_classification

# Build a classification task using 3 informative features
X = [ df[df['is_train']==True].drop(['serial','Mathematics','is_train'],axis=1) for df in df_breakdown ]
y = [ df[df['is_train']==True]['Mathematics'] for df in df_breakdown ]

# Create the RFE object and compute a cross-validated score.
svc = SVC(kernel="linear")
# The "accuracy" scoring is proportional to the number of correct
# classifications
rfecv = [RFECV(estimator=svc, step=1, cv=StratifiedKFold(cv_inst, 2),
              scoring='accuracy') for cv_inst in y]
cv = [ rfe_inst.fit(features, find) for features,find,rfe_inst in zip(X,y,rfecv)]


# Plot number of features VS. cross-validation scores
import matplotlib.pyplot as plt
for rfe_inst in rfecv:
    print("Optimal number of features : %d" % rfe_inst.n_features_)
    plt.figure()
    plt.xlabel("Number of features selected")
    plt.ylabel("Cross validation score (nb of correct classifications)")
    plt.plot(range(1, len(rfe_inst.grid_scores_) + 1), rfe_inst.grid_scores_)
    plt.show()
```

```python
pca = [ PCA(n_components=rfe_inst.n_features_) for rfe_inst in rfecv]
X_r = [ pca_inst[0].fit_transform(pca_inst[1]) for pca_inst in zip(pca,X)]
X_r
```

```python
#Now train a classifier and look at results
clf = SVC(kernel="linear")
classifiers = [clf.fit(tr[features].drop('Mathematics',axis=1), tr['Mathematics']) for tr,features in zip(train,featuresets)]
preds = [ cl.predict(ts[fs].drop('Mathematics',axis=1)) for cl,ts,fs in zip(classifiers,test,featuresets)]
preds
```
